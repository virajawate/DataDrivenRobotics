```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
n = 256
w = np.exp(-1j * 2 * np.pi / n)
```

```python
DFT = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        DFT[i, j] = w ** (i*j)

DFT = np.real(DFT)
# compexity O(n^2)
```

```python
plt.imshow(DFT)
plt.show()
```

```python
plt.hist(DFT)
plt.show()
```

```python
J, K = np.meshgrid(np.arange(n*0.5), np.arange(n*0.5))
DFT2 = np.power(w, J*K)
DFT2 = np.real(DFT2)
```

```python
plt.imshow(DFT2)
plt.show()
```

Discrete Fourier transform converts a finite sequence of equally-spaced samples of a function into a same-length sequence of equally-spaced samples of the discrete-time Fourier transform, which is a complex-valued function of frequency.
