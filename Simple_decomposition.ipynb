```python
import os
import numpy as np
import numpy.linalg as algo
import matplotlib.pyplot as plt
from matplotlib.image import imread

plt.rcParams['figure.figsize'] = [15, 8]
```

```python
raw = imread('./data/dog.jpg')
img = np.mean(raw, -1)
```

```python
im_read = plt.imshow(img)
im_read.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
# Performing Sigular Value Decomposition (SVD)
U, S, VT = algo.svd(img, full_matrices=False)
S = np.diag(S)
```

```python
from ipywidgets import widgets, interact

def singular_value(r):
    X_approx = U[:, :r] @ S[0:r, :r] @ VT[:r, :]
    im_svd = plt.imshow(X_approx, cmap='gray')
    plt.axis('off')
    plt.title(f'r = {str(r)}')
    plt.show()

# Assuming X, U, S, VT are defined before this code snippet
interact(singular_value, r=widgets.IntSlider(min=0, max=min(img.shape), step=1, value=5))
```

```python
plt.subplot(2,1,1)
plt.semilogy(np.diag(S))
plt.title('Singular_Value')
plt.show()
plt.subplot(2,1,2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title("Singular Values: Cumulative Sum")
plt.show()
```
